```python
import os
from PyPDF2 import PdfReader

# Diretrio contendo os PDFs
pdf_dir = './literatura/sensoriamento_remoto'
pdf_files = [f for f in os.listdir(pdf_dir) if f.endswith('.pdf')]

# Funo para extrair texto de um PDF
def extract_text_from_pdf(pdf_path):
    with open(pdf_path, 'rb') as file:
        reader = PdfReader(file)
        text = ''
        for page_num in range(len(reader.pages)):
            text += reader.pages[page_num].extract_text() or ''
        return text

# Extrair texto de todos os PDFs
pdf_texts = [extract_text_from_pdf(os.path.join(pdf_dir, pdf_file)) for pdf_file in pdf_files]
```

```python
import os
from groq import Groq
from llama_index.core import ServiceContext
from llama_index.llms.monsterapi import MonsterLLM
llm = MonsterLLM(model="meta-llama/Meta-Llama-3-8B-Instruct", temperature=0.5, context_window=2024)
service_context = ServiceContext.from_defaults(
    chunk_size=2024, llm=llm, embed_model="local:BAAI/bge-small-en-v1.5"
)
```

```python
from llama_index.core import VectorStoreIndex

index = VectorStoreIndex.from_documents(pdf_texts, service_context=service_context)
query_engine = index.as_query_engine()
```
